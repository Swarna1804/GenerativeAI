```python
from transformers import AutoTokenizer,AutoModelForSequenceClassification, pipeline
```

```python
model_name="distilbert-base-uncased-finetuned-sst-2-english"
```

```python
tokenizer=AutoTokenizer.from_pretrained(model_name)
model=AutoModelForSequenceClassification.from_pretrained(model_name)
```

```python
sentiment_analyzer=pipeline("sentiment-analysis",model=model,tokenizer=tokenizer,device=-1)
```

```python
texts=[
    "I'm good but not happy.",
    "Today is a good day.",
    "It's okay,not great but acceptable."
]
```

```python
results=sentiment_analyzer(texts)
```

```python
for text, result in zip(texts, results):
  print(f"Text: {text}\n  Sentiment:{result['label']},confidence:{result['score']:.2f}\n")
  # Use result instead of results to access the current dictionary in the loop
```

```python
from transformers import AutoTokenizer, AutoModelForSequenceClassification, pipeline

# Using a different model
model_name = "roberta-base-openai-detector"  # Updated model name
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForSequenceClassification.from_pretrained(model_name)

# Setting up the sentiment analysis pipeline
sentiment_analyzer = pipeline("sentiment-analysis", model=model, tokenizer=tokenizer, device=-1)

# Test texts
texts = [
    "I'm good but not happy.",
    "Today is a good day.",
    "It's okay, not great but acceptable."
]

# Analyzing sentiment
results = sentiment_analyzer(texts)

# Displaying results
for text, result in zip(texts, results):
    print(f"Text: {text}\n  Sentiment: {result['label']}, Confidence: {result['score']:.2f}\n")
```
